## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = "SELECT ConvertedComp FROM master"
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(data['ConvertedComp'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

import matplotlib.pyplot as plt

# Plotting the box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data['Age'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = "SELECT Age, WorkWeekHrs FROM master"
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['WorkWeekHrs'], alpha=0.5, color='skyblue')
plt.title('Scatter Plot of Age vs WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = "SELECT Age, WorkWeekHrs, CodeRevHrs FROM master"
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Define bubble size
bubble_size = data['Age'] * 2  # Adjust size scaling as needed

# Plotting the bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(data['WorkWeekHrs'], data['CodeRevHrs'], s=bubble_size, alpha=0.5, color='skyblue')
plt.title('Bubble Plot of WorkWeekHrs vs CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.grid(True)
plt.show()

```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = """
        SELECT Age, WorkWeekHrs, CodeRevHrs
        FROM master
        WHERE Age >= 30 AND Age <= 35
        """
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Group data by Age and calculate the median WorkWeekHrs and CodeRevHrs
grouped_data = data.groupby('Age').median()

# Select only the age group 30 to 35
grouped_data = grouped_data.loc[30:35]

# Plotting the stacked chart
grouped_data[['WorkWeekHrs', 'CodeRevHrs']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)
plt.legend(['WorkWeekHrs', 'CodeRevHrs'])
plt.grid(axis='y')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = """
        SELECT Age, ConvertedComp
        FROM master
        WHERE Age >= 45 AND Age <= 60
        """
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Group data by Age and calculate the median ConvertedComp
grouped_data = data.groupby('Age').median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
plt.plot(grouped_data.index, grouped_data['ConvertedComp'], marker='o', color='skyblue', linestyle='-')
plt.title('Median ConvertedComp for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(range(45, 61, 1))
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query data from the database
query = "SELECT MainBranch, COUNT(*) AS count FROM master GROUP BY MainBranch ORDER BY count DESC"
data = pd.read_sql_query(query, conn)

# Close the database connection
conn.close()

# Plotting the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(data['MainBranch'], data['count'], color='skyblue')
plt.title('Number of Respondents by Main Branch')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.gca().invert_yaxis()  # Invert y-axis to display the highest count at the top
plt.grid(axis='x')
plt.show()
```
